### Feature Generation

Converting Date_Of_Order and Merchant_Registration_Date as datetime64 for both training and testing data

```python
trainingData['Date_of_Order'] = trainingData['Date_of_Order'].astype('datetime64[ns]')
trainingData['Merchant_Registration_Date'] = trainingData['Merchant_Registration_Date'].astype('datetime64[ns]')
trainingData.head()
```

```python
testingData['Date_of_Order'] = testingData['Date_of_Order'].astype('datetime64[ns]')
testingData['Merchant_Registration_Date'] = testingData['Merchant_Registration_Date'].astype('datetime64[ns]')
testingData.head()
```

Creating a new_feature Difference_Date by substracting Date_of_order and Merchant_registration_date

```python
trainingData['difference_dates']=(trainingData["Date_of_Order"])- ( trainingData["Merchant_Registration_Date"])
trainingData.head()
```

```python
testingData['difference_dates']=(testingData["Date_of_Order"])- ( testingData["Merchant_Registration_Date"])
testingData.head()
```

Dividing difference_dates by day we get the difference

```python
trainingData["difference_dates"]=trainingData["difference_dates"]/np.timedelta64(1, "D")
trainingData.head()
```

```python
testingData["difference_dates"]=testingData["difference_dates"]/np.timedelta64(1, "D")
testingData.head()
```

### Feature dummification

Getting dummies of training and testing datasets

```python
trainingData=pd.get_dummies(trainingData)
trainingData.head()
```

```python
testingData=pd.get_dummies(testingData)
testingData.head()
```

```python
#Spliting between numerical and categorical attributes in trainingData
cat_attr=list(trainingData.select_dtypes("category").columns)
num_attr = list(trainingData.columns.difference(cat_attr))
```

```python
num_attr
```

### Making our ML pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn import preprocessing
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
```

```python
categoricalTransformer=Pipeline(steps=[
                                      ('imputer',SimpleImputer(strategy='most_frequent')),
                                      ('onehot',OneHotEncoder(handle_unknown='ignore'))])
```

```python
numericTransformer=Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())])
```

```python
preprocessor=ColumnTransformer( transformers=[
        ('num', numericTransformer, num_attr),
        ('cat', categoricalTransformer, cat_attr)])
preprocessor
```

### Building Our Model

```python
#Splitting the dataset into training and testing datasets:
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(trainingData,Y,test_size=0.2,random_state=45)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### 1.Using Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lrmodel=Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('classifier',LogisticRegression())
])
```

```python
lrmodel.fit(x_train,y_train)
```

```python
trainPrediction=lrmodel.predict(x_train)
lrmodel.score(x_train,y_train)
```

```python
testPrediction=lrmodel.predict(x_test)
lrmodel.score(x_test,y_test)
```

```python
#getting accuracy and precision data out of our training data
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('The train score accuracy is {}'.format(accuracy_score(y_train,trainPrediction)))
print('The precision score for training is {}'.format(precision_score(y_train,trainPrediction)))
print('The recall score for training is {}'.format(recall_score(y_train,trainPrediction)))
print('The F1 score for training is {}'.format(f1_score(y_train,trainPrediction)))
```

```python
print('The test score accuracy is {}'.format(accuracy_score(y_test,testPrediction)))
print('The precision score for testing is {}'.format(precision_score(y_test,testPrediction)))
print('The recall score for testing is {}'.format(recall_score(y_test,testPrediction)))
print('The F1 score for testing is {}'.format(f1_score(y_test,testPrediction)))
```

```python
#Making a prediction based on our model on the test data
regressionValues=lrmodel.predict(testingData)
regressionValues
```

```python
fraudsterPrediction=pd.DataFrame({
    'Merchant_ID':testingData['Merchant_ID'],
    'Fraudster':regressionValues
})
#fraudsterPrediction.to_csv('logisticRegressionResults.csv',index=False)
```

```python
#Plotting the predicted fraudsters in our testingData using linear regression model
sns.countplot(x='Fraudster',data=fraudsterPrediction,hue='Fraudster')
```

```python
fraudsters=len(fraudsterPrediction[fraudsterPrediction.Fraudster==1])
TotalSample=len(fraudsterPrediction.Fraudster)
merchants=TotalSample-fraudsters
fraudstersPercentage=(fraudsters/TotalSample)*100
merchantsPercentage=(merchants/TotalSample)*100
print('The percentage of fraudsters in the predicted output is {}%'.format(fraudstersPercentage))
print('The percentage of merchants in the predicted output is {}%'.format(merchantsPercentage))
```

#### 2.Using Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
%%time
svcModel = Pipeline(steps=[('preprocessor', preprocessor),
                      ('classifier', SVC())])
svcModel.set_params(classifier__kernel='linear',classifier__C=1,
                   classifier__random_state=123)

svcModel.fit(x_train,y_train)
```

```python
trainPrediction=svcModel.predict(x_train)
svcModel.score(x_train,y_train)
```

```python
testPrediction=svcModel.predict(x_test)
svcModel.score(x_test,y_test)
```

```python
print('The train score accuracy is {}'.format(accuracy_score(y_train,trainPrediction)))
print('The precision score for training is {}'.format(precision_score(y_train,trainPrediction)))
print('The recall score for training is {}'.format(recall_score(y_train,trainPrediction)))
print('The F1 score for training is {}'.format(f1_score(y_train,trainPrediction)))
```

```python
print('The test score accuracy is {}'.format(accuracy_score(y_test,testPrediction)))
print('The precision score for testing is {}'.format(precision_score(y_test,testPrediction)))
print('The recall score for testing is {}'.format(recall_score(y_test,testPrediction)))
print('The F1 score for testing is {}'.format(f1_score(y_test,testPrediction)))
```
